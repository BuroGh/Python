```python
# 4> Describe how dictionaries store data?

'''Dictionaries in Python store data as key-value pairs, where each key is unique, and it maps to a corresponding value. This structure allows for efficient lookups, insertions, and deletions.

Underlying Mechanism: Hash Tables
i) Dictionaries use hash tables as their underlying data structure.
ii) Each key is passed through a hash function, which generates a unique integer (called the hash value or hash code) for the key.
iii) The hash value determines the location (bucket) where the key-value pair is stored in memory.

Steps of Storage:

1)Key Hashing:
When a key is added, it is passed through a hash function (hash() in Python) to generate a hash value.

2)Index Calculation:
The hash value is used to calculate an index in the hash table where the key-value pair is stored.

3)Storing Key-Value Pair:
The key and value are stored together in the calculated bucket.

Retrieving Data:
When accessing a value by its key, the key is hashed again, and the hash value is used to locate the bucket containing the key-value pair.



Example:

# Creating a dictionary
my_dict = {"name": "Alice", "age": 25, "city": "New York"}

# Adding a key-value pair
my_dict["country"] = "USA"

# Accessing a value
print(my_dict["name"])  # Output: Alice

# Modifying a value
my_dict["age"] = 26

# Removing a key-value pair
del my_dict["city"]

# Final dictionary
print(my_dict)  # Output: {'name': 'Alice', 'age': 26, 'country': 'USA'}

```

```python
#8> What is a hash table, and how does it relate to dictionaries in Python?

'''A hash table is a data structure that stores data in an associative manner. It uses a hash function to compute an index (or hash value) for a given key, which determines where the corresponding value will be stored in memory. Hash tables allow for efficient data retrieval and storage, typically with an average time complexity of O(1) for operations like lookups, insertions, and deletions.

Components of a Hash Table

Keys:
Unique identifiers used to store and retrieve values.

Values:
Data associated with the keys.

Buckets:
Logical storage locations in the hash table where key-value pairs are placed.

Hash Function:
A function that maps keys to specific buckets (indices) in the table.




Python's dictionaries are implemented using hash tables under the hood. Here’s how they work:

1. Key Hashing
When you add a key-value pair to a dictionary, the key is passed through a hash function (hash() in Python).
The hash function generates an integer (hash value) that determines which bucket will store the key-value pair.

2. Index Calculation
The hash value is further processed (e.g., using modulo operation) to map it to a specific index in the dictionary's hash table.

3. Storage
The key-value pair is stored in the bucket corresponding to the calculated index.

4. Lookup
To retrieve a value, the key is hashed again, and the hash value is used to locate the bucket where the key-value pair is stored.
```

```python
#17> Describe the time complexity of accessing elements in a dictionary.

'''Time Complexity of Accessing Elements in a Dictionary
In Python, dictionaries are implemented using hash tables, which allow for very efficient lookups, insertions, and deletions. The time complexity of accessing elements in a dictionary is typically O(1), also known as constant time.

This means that no matter how large the dictionary is, the time it takes to access an element (given its key) does not increase significantly with the size of the dictionary.

Why is the Time Complexity O(1)?
Hashing: When you attempt to access an element in a dictionary, Python first applies a hash function to the key. This hash function converts the key into a hash value (an integer), which corresponds to a specific location or "bucket" in the dictionary's underlying structure.
Direct Access: The dictionary then directly accesses the bucket corresponding to the hash value and retrieves the value associated with the key. Since hashing allows for direct indexing, the lookup operation is very fast.
For most cases, this operation takes constant time because it doesn't depend on the size of the dictionary.

Worst-Case Time Complexity: O(n)
In some rare situations, the time complexity can degrade to O(n) (linear time), such as when there are many hash collisions. A collision happens when two different keys produce the same hash value. In this case, multiple keys are stored in the same "bucket," and the dictionary may have to search through a list of values to find the correct one.

However, Python's dictionary implementation uses techniques like open addressing and dynamic resizing to minimize the occurrence of collisions, making the average time complexity close to O(1).

Example of Dictionary Access
python
Copy
Edit
my_dict = {'apple': 5, 'banana': 10, 'cherry': 15}

# Accessing the value for a given key
apple_count = my_dict['apple']  # O(1) time complexity

print(apple_count)  # Output: 5
In this example, accessing my_dict['apple'] takes constant time, regardless of the size of the dictionary.

```

```python
#18> In what situations are lists preferred over dictionaries?

'''While both lists and dictionaries are commonly used data structures in Python, they are suited to different kinds of tasks. Here are the main situations where lists are preferred over dictionaries:

1. When Order Matters (Indexed Access)
Lists maintain the order of elements, meaning the elements are stored in the order in which they are added. You can access elements by their index (position in the list).
Dictionaries, on the other hand, are unordered collections (although they maintain insertion order starting from Python 3.7, this does not mean they are based on indices).
Use Case:
When you need to store a sequence of items and access them based on their position rather than their key, a list is ideal.

python
Copy
Edit
# Example where order matters
my_list = [10, 20, 30, 40]
print(my_list[2])  # Output: 30
2. When You Need a Sequence of Items (Without Unique Keys)
Lists are designed for handling collections of ordered items, where the same element can appear multiple times. You can have duplicates in a list.
Dictionaries require unique keys, meaning each key can only map to one value, which prevents duplicates based on the key.
Use Case:
If you want to store a collection of repeated values or don't need to map each element to a unique key, lists are the way to go.

python
Copy
Edit
# Example where duplicates are allowed
my_list = [1, 2, 2, 3, 3, 3]
print(my_list.count(3))  # Output: 3
3. When You Need to Iterate Over Items Sequentially
Lists are great for tasks that require you to process items sequentially. You can iterate over a list in order, one item at a time, or use built-in functions like map(), filter(), and list comprehensions.
Dictionaries are typically used when you need to access data by keys. Iterating over dictionaries usually happens over keys or values, and is more suited for situations where you need to perform operations on key-value pairs rather than the sequence of elements.
Use Case:
If your task requires iterating through elements in order or applying transformations to each element of a collection, a list is preferred.

python
Copy
Edit
# Example of iterating through a list
my_list = [1, 2, 3, 4]
for item in my_list:
    print(item)
4. When You Don't Need Key-Value Pairs
Lists store data as simple sequences of values.
Dictionaries store data as key-value pairs, which are useful when you need to map one piece of information to another (e.g., name to age). If you don't need this mapping functionality, a list is more appropriate.
Use Case:
If you are simply storing and accessing values without any need to associate them with unique keys, a list is sufficient.

python
Copy
Edit
# Example where key-value pair is not needed
my_list = [100, 200, 300]
5. When Memory Efficiency is a Concern for Simple Collections
Lists are typically more memory efficient than dictionaries when dealing with simple collections, especially when you don't need the additional overhead of keys.
Dictionaries use a hash table under the hood, which introduces some extra memory overhead due to the storage of keys and values.
Use Case:
When dealing with a large collection of items that do not need to be mapped to unique keys and you care about memory efficiency, lists are the preferred choice.

```

```python
#19> Why are dictionaries considered unordered, and how does that affect data retrieval?

'''Dictionaries in Python are considered unordered collections because the items stored within a dictionary are not guaranteed to maintain any specific order before Python version 3.7.

In Python pre-3.7, the internal structure used to store key-value pairs did not maintain any order of insertion. The dictionary was implemented using a hash table, where the keys were hashed and mapped to specific locations in memory, but this did not guarantee that elements would be stored or retrieved in the same order in which they were added.

However, starting with Python 3.7 and onwards, dictionaries preserve insertion order. This means that the order in which key-value pairs are inserted into a dictionary is maintained when you iterate over the dictionary. This feature became an official part of the language specification in Python 3.7, but dictionaries are still technically unordered in the sense that they are not based on an indexed collection like lists.

Impact of Dictionary Order on Data Retrieval
Even though dictionaries in Python 3.7+ maintain insertion order, order is not intrinsic to their primary functionality. This is because dictionaries are designed to provide efficient key-based lookups, not positional access or ordered access.

Here’s how the concept of being "unordered" affects data retrieval in dictionaries:

1. Data Retrieval by Key (O(1) Time Complexity)
Dictionaries use hash tables for fast lookups, which means accessing values by their keys is efficient and done in constant time (O(1)) on average.
The key-value pair is retrieved by applying a hash function to the key and looking up its corresponding value. This process does not depend on the order of the keys.
python
Copy
Edit
my_dict = {'a': 1, 'b': 2, 'c': 3}
print(my_dict['b'])  # Output: 2
In this example, the value associated with the key 'b' is retrieved in constant time (O(1)), regardless of the order in which 'a', 'b', and 'c' were inserted.

2. Iteration Order (Preserved in Python 3.7+)
Although dictionaries are unordered in terms of direct indexing, Python 3.7+ preserves insertion order during iteration. This means that when you iterate over the dictionary, the keys (and values) will be returned in the same order in which they were added to the dictionary.
python
Copy
Edit
# Python 3.7+ maintains insertion order
my_dict = {'a': 1, 'b': 2, 'c': 3}
for key, value in my_dict.items():
    print(key, value)
Output:

css
Copy
Edit
a 1
b 2
c 3
In this case, the dictionary’s insertion order is preserved during iteration.

3. No Index-Based Access
Dictionaries do not support indexing like lists do. You cannot access dictionary items based on their position (like my_dict[0] or my_dict[1]), because they are unordered collections. You must retrieve items by their key.
python
Copy
Edit
# This will raise a TypeError, because dictionaries don't support indexing
my_dict[0]  # TypeError: 'dict' object is not subscriptable
4. Searching and Filtering
Searching for items in a dictionary is typically done by key lookup (i.e., using key in dict). However, the insertion order of items doesn't affect this operation.
python
Copy
Edit
# Checking if a key exists
print('a' in my_dict)  # Output: True
5. No Guarantee of Ordering Before Python 3.7
Before Python 3.7, dictionaries did not maintain order at all, and the order in which you added key-value pairs was irrelevant. If you tried to iterate over the dictionary, you couldn’t predict the order in which the items would be returned.
```

```python
#20> Explain the difference between a list and a dictionary in terms of data retrieval.

'''Both lists and dictionaries are fundamental data structures in Python, but they differ significantly in how data is stored and retrieved. Let's explore the key differences in terms of data retrieval:

1. Access Method:
List:

Lists are indexed collections. Data in a list is accessed using an integer index, starting from 0.

You retrieve data from a list by specifying the index position of the element.

python
Copy
Edit
my_list = [10, 20, 30, 40]
print(my_list[2])  # Output: 30
In this example, the element 30 is retrieved using the index 2.

Dictionary:

Dictionaries are key-value pairs. Data in a dictionary is accessed using a unique key (which can be a string, number, or other immutable types).

You retrieve data from a dictionary by specifying the key, not an index.

python
Copy
Edit
my_dict = {'apple': 5, 'banana': 10, 'cherry': 15}
print(my_dict['banana'])  # Output: 10
In this case, the value 10 is retrieved using the key 'banana'.

2. Data Type for Access:
List:
In lists, access is done using integers (index positions).
The order of elements in the list is significant, and you access elements based on their position in the sequence.
Dictionary:
In dictionaries, access is done using keys (which can be strings, numbers, or tuples). Each key maps to a corresponding value.
The order of elements is not as important for retrieval since data is accessed by key, not by position.
3. Time Complexity:
List:

Accessing an element by index in a list is an O(1) operation (constant time).

However, if you are looking for an element by value (i.e., if the value is not known and you need to search the entire list), it takes O(n) time, where n is the number of elements in the list.

python
Copy
Edit
my_list = [10, 20, 30, 40]
print(30 in my_list)  # O(n) time complexity to check if the value is in the list
Dictionary:

Accessing an element by key in a dictionary is an O(1) operation (constant time) on average, because dictionaries use a hash table for efficient key-based lookups.

Searching by key is fast, but searching by value (without the key) would take O(n) time since you would need to iterate through all the key-value pairs to find a matching value.

python
Copy
Edit
my_dict = {'apple': 5, 'banana': 10, 'cherry': 15}
print('banana' in my_dict)  # O(1) time complexity to check if the key exists
4. Order of Data Retrieval:
List:

Lists are ordered collections, so elements are accessed in the exact order in which they were added (i.e., by their index).

If you iterate through a list, the order will always be preserved.

python
Copy
Edit
my_list = [10, 20, 30]
for item in my_list:
    print(item)
# Output:
# 10
# 20
# 30
Dictionary:

Dictionaries are unordered collections in terms of how they are stored internally. Before Python 3.7, there was no guarantee that the order of the key-value pairs would be preserved.

Starting with Python 3.7, dictionaries maintain insertion order, but this does not affect the way elements are retrieved (you access them by key, not position).

python
Copy
Edit
my_dict = {'apple': 5, 'banana': 10, 'cherry': 15}
for key, value in my_dict.items():
    print(key, value)
# Output (Python 3.7+):
# apple 5
# banana 10
# cherry 15
5. Use Cases for Access:
List:
Lists are ideal when you need ordered collections of items where the position of each element matters. They are often used when you need to iterate over a collection in sequence or access items by index.
Example: Storing a sequence of numbers or strings where the order is important.
Dictionary:
Dictionaries are ideal when you need to store key-value pairs and retrieve values based on a unique key. They provide quick lookups based on keys and are used when you need to associate a specific piece of information with another.
Example: Storing information like employee IDs and their names, or mapping cities to their populations.
```
